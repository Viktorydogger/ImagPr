```python
import cv2
import numpy as np
import math as m

# load image as grayscale
img = cv2.imread('pic2.jpeg')

gray = cv2.cvtColor(img, cv2.COLOR_RGBA2GRAY)

cv2.imshow('ROI',gray)
cv2.waitKey()
# threshold input image using otsu thresholding as mask and refine with morphology
ret, mask = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY_INV+cv2.THRESH_OTSU) 
kernel = np.ones((29,29), np.uint8)
mask = cv2.morphologyEx(mask, cv2.MORPH_CLOSE, kernel)
mask = cv2.morphologyEx(mask, cv2.MORPH_OPEN, kernel)

# put thresh into 
result = img.copy()
result = cv2.cvtColor(result, cv2.COLOR_BGR2BGRA)
result[:, :, 3] = mask

# save resulting masked image
cv2.imwrite('retina_masked.png', result)
```

```python
image = cv2.imread('wom.jpg')
result = image.copy()
gray = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
thresh = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)[1]
horizontal_kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (1,1))
detect_horizontal = cv2.morphologyEx(thresh, cv2.MORPH_OPEN, horizontal_kernel, iterations=2)
cnts = cv2.findContours(detect_horizontal, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
cnts = cnts[0] if len(cnts) == 2 else cnts[1]
for c in cnts:
    cv2.drawContours(result, [c], -1, (36,255,12), 2)

cv2.imshow('ROI',result)
cv2.waitKey()
```

```python
image = cv2.imread("wom.jpg")
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
blurred = cv2.GaussianBlur(gray, (3, 3), 0)
edged = cv2.Canny(blurred, 10, 100)

# define a (3, 3) structuring element
kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (3, 3))

# apply the dilation operation to the edged image
dilate = cv2.dilate(edged, kernel, iterations=1)

# find the contours in the dilated image
contours, _ = cv2.findContours(dilate, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
image_copy = image.copy()
# draw the contours on a copy of the original image
cv2.drawContours(image_copy, contours, -1, (0, 255, 0), 2)
print(len(contours), "objects were found in this image.")

cv2.imshow("Dilated image", dilate)
cv2.imshow("contours", image_copy)
cv2.waitKey(0)
```

```python
import cv2
import numpy as np
# достаём страничку
image = cv2.imread('pic2.jpeg')
imageold=cv2.imread('pic2.jpeg')
fig = np.zeros((1,5))
iost=0
S = 0
Sim = image.shape[0]*image.shape[1]
for i in range(100):
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    mask = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)[1]
    cnts = cv2.findContours(mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    cnts = cnts[0] if len(cnts) == 2 else cnts[1]
    cnts = sorted(cnts, key=cv2.contourArea, reverse=True)
# вырезаем самую большую область
    for c in cnts:
        x,y,w,h = cv2.boundingRect(c)
        fig = np.append(fig,[[x**2+y**2,x,y,w,h]],axis = 0)
        ROI1 = image[y:y+h, x:x+w]
        S += h*w
        if S/Sim > 0.8: 
            iost += 1
            break
        break
    if iost: break
    for c in cnts:
        cv2.drawContours(image, [c], -1, (255,255,255),-1)
        break
fig = np.delete(fig,0,axis=0)
fig1 = fig[fig[:,0].argsort()][fig[:,2].argsort()]
fig1 = fig1.astype(int)

for i in range(iost):
    x,y,w,h = fig1[i,1],fig1[i,2],fig1[i,3],fig1[i,4]
    ROI = imageold[y:y+h, x:x+w]
    cv2.imshow('ROI',ROI)
    cv2.waitKey()
```
